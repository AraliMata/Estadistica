## D. Preparación de los datos

### Selección de conjunto de datos

**Variables no utilizadas:**
* Se decidió no utilizar las variables **salary** y **salary_currency** ya que es necesario que la variable objetivo (salary_in_usd) se encuentre en las mismas unidades y estas dos variables no proporcionan nueva información.

**Variables utilizadas:**
* Se decidió utilizar las variables **work_year**,**experience_level**, **employment_type**, **job_title**, **employee_residence**, **remote_ratio**,**company_location**, **company_size** debido a que, como se observa en el punto anterior, tienen cierta influencia en el salario.

**Variable objetivo**
* Salario en dolares (salary_in_usd)

### Transformación de datos categóricos.


*   Para la transformación de las variables ordinales se va a utilizar la clase OrdinalEncoder(), ya que toma los valores de las variables y los convierte en numéricos.
*   Para la transformación de las variables nominales se van a obtener las variables dummies de estas.



**Variables ordinales**

* Remote ratio
* Experience level
* Company size

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
ordinal = pd.DataFrame()
encoder.fit(data[["experience_level","company_size"]])
ordinal[["experience_level","company_size"]] = encoder.transform(data[["experience_level","company_size"]])
ordinal['remote_ratio'] = data["remote_ratio"]

ordinal
```

```python
ordinal.count()
```

**Variables nominales**

* work_year
* employment_type
* job_title
* employee_residence
* company_location

```python
work_year = pd.get_dummies(data.work_year, prefix='year')
employment_type = pd.get_dummies(data.employment_type, prefix='employment')
job_title = pd.get_dummies(data.job_title, prefix='job_title')
employee_residence = pd.get_dummies(data.employee_residence, prefix='employee_residence')
company_location = pd.get_dummies(data.company_location, prefix='company_location')

X = pd.concat([ordinal, work_year, employment_type, job_title, employee_residence, company_location], axis=1)
y = data["salary_in_usd"]
data = pd.concat([y, X], axis=1)
data
```

```python
data.count()
```

### Manejo de datos duplicados y atípicos.

Debido a que solo se encontraron 8 outliers y 42 duplicados, se decide eliminarlos.  

```python
std_value= data["salary_in_usd"].std()
data = data.loc[np.abs((data["salary_in_usd"] - data["salary_in_usd"].mean()) / std_value)  < 3]
```

```python
data = data.loc[data.duplicated() == False]
```

### Escalamiento, normalización y discretización

En este caso no es necesario **discretizar** los datos, ya que solo se tiene una variable numérica y se trata de la variable objetivo (salary_in_usd), la cual es discreta.

Debido a que la variable numérica salary_in_usd no tiene una forma normal, esta se va a **normalizar** para que sea apta para utilizarla en la propuesta de algún modelo. El resto de las variables se va a dejar sin normalizar ya que la mayoría se encuentra separada en variables dummies que solo toman valores de 0 y 1.

```python
"""%%R
install.packages("MASS")"""
```

```python
%%R -i data
library(MASS)
salary = data$salary_in_usd
salary_in_usd_box <- boxcox(salary~1)
salary_in_usd_V = salary_in_usd_box$x[which.max(salary_in_usd_box$y)]
salary_in_usd_V
```

```python
%%R
salary_in_usd_normalized = (salary ^ salary_in_usd_V - 1) / salary_in_usd_V
data$salary_in_usd = salary_in_usd_normalized
hist(data$salary_in_usd)
```

También se realizará un **escalamiento** de los datos para que todas las variables estén en la misma escala, esto con un escalamiento estándar.

```python
%%R
z_norm<-function(x){(x-mean(x))/sd(x)}
M1=as.data.frame(lapply(data,z_norm))
```
